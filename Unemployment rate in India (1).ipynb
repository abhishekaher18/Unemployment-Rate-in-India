```python
sns.heatmap(df.corr(),cmap='YlGnBu',annot=True)
```

```python
sns.countplot(data = df, x = 'Region', palette='CMRmap')
plt.xticks(rotation=90)
plt.title('Unemployment', color = 'black')
plt.show()
```

```python
labels = df['Region'].value_counts().index
sizes = df['Region'].value_counts()
colors= ['b','g','r','m','y','w','c']
plt.figure(figsize = (8,8))
plt.pie(sizes,labels = labels, rotatelabels = False, autopct ='%1.1f%%',colors=colors,shadow = True, startangle = 0)
plt.title('Unemployment by Region',color = 'g',fontsize = 20)
plt.legend(bbox_to_anchor=(1.05,1),title = 'Region:')
plt.show()
```

```python
sns.histplot(data = df, x ='Area', hue='Area',shrink =.5,alpha=1)
```

```python
labels = df['Area'].value_counts().index
sizes = df['Area'].value_counts()
colors= ['m','c']
plt.figure(figsize = (8,8))
plt.pie(sizes,labels = labels, rotatelabels = False, autopct ='%1.1f%%',colors=colors,shadow = True, startangle = 0)
plt.title('Unemployment by Area',color = 'g',fontsize = 20)
plt.legend(bbox_to_anchor=(1.05,1),title = 'Area:')
plt.show()
```

```python
plt.plot(df.Region.value_counts(),'o:r',ms=10,mec='k')
plt.xlabel('Region',fontsize=10,color='m')
plt.ylabel('Count',fontsize=10,color='m')
plt.title("Region",color='g')
plt.xticks(rotation=90)
```

```python
plt.plot(df.Area.value_counts(),'o:r',ms=10,mec='k')
plt.xlabel('Area',fontsize = 10,color = 'm')
plt.title("Area",color='g')
```
